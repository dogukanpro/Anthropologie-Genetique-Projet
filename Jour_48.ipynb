Sujet : Comparaison de deux listes avec zip ()

```python
#Initialisation
ancetre = "ATGCGT"
moderne = "ATGGGT"
compteur_mutations = 0
```

La ou la comparaison s'effectue

```python
for base_anc, base_mod in zip(ancetre, moderne):
    if base_anc != base_mod:
        print(f"Mutation trouvée : {base_anc} est devenu {base_mod}")
        compteur_mutations += 1
    else:
        print(f"Identique : {base_anc}")
```

Affichage

```python
print(f"\nNombre total de mutations : {compteur_mutations}")
```
