Création algorithme qui peut stocker des informations différentes pour différents peuple, dans un fichier texte

```python
def ajouter_historique(peuple, score):
    nom_f = "historique_analyses.txt"
    
    with open(nom_f, "a") as f:
        f.write(f"Peuple: {peuple} | Score: {score}%\n")
    
    print(f"\n Entrée ajoutée pour les {peuple}.")
```

Affichage

```python
ajouter_historique("Celtes", 24.5)
ajouter_historique("Vikings", 12.8)
ajouter_historique("Romains", 45.2)
```
