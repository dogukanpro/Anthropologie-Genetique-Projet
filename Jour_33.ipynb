Calcule de l'écart génétique entre deux populations

```python
def calculer_distance(peuple_A, freq_A, peuple_B, freq_B):
    distance = freq_A - freq_B
    print(f"Analyse entre {peuple_A} et {peuple_B}")
    print(f"Écart génétique : {distance:.4f}")
    
    if distance < 0.02:
        return "Ces deux peuples sont génétiquement très proches."
    else:
        return "Ces deux peuples montrent une divergence significative."
```

Par exemple mettons que les suédois ont une frequence allelique pour "x", de 74% et que les anciens fermiers en avaient 12%

```python
verdict = calculer_distance("Suédois", 0.74, "Anciens Fermiers", 0.12)
print(verdict)
```
