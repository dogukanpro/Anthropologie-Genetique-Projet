```python
adn_brut = ["atgc", "atgcgtac", "at", "ggccata", "c"]
```

Boucle for + if

```python
for s in adn_brut : 
    majuscule= s.upper()
    longueur = len(s)

if longueur >= 5 :
    print(f"Séquence {majuscule} : OK (Longueur : {longueur})")
else :
    print(f"Séquence {majuscule} : Erreur (Trop courte)")
```
