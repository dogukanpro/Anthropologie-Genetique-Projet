Initialisation

```python
population1 = "ATGCGT"
population2 = "ATGGGT"
```

```python
Utilisation de la boucle for
```

```python
def calculer_distance(seq1, seq2):
    score = 0
    
    for i in range(len(seq1)):
        if seq1[i] != seq2[i]:
            score = score + 1
    return score
```

Comparaison entre les deux populations + affichage

```python
diff = calculer_distance(population1, population2)
print(f"Nombre de mutations de différence : {diff}")
```
