Initialisation

```python
profil_genetique = {
"Groupe_Sanguin": "O", "Persistance_Lactase": "Non","Couleur_Yeux": "Bleu","Mutation_Albinisme": "Non"
}
marqueur_recherche = "Persistance_Lactase"
```

Utilisation de la boucle if + affichage

```python
if marqueur_recherche in profil_genetique:
    resultat = profil_genetique[marqueur_recherche]
    print(f"Analyse du gène {marqueur_recherche} terminée.")
    print(f"Résultat pour ce fossil : {resultat}")
else:
    print(f"Erreur, le gène {marqueur_recherche} n'a pas pu être séquencé")
```
