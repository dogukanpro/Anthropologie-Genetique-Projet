Initialisation

```python
with open("brut.txt", "w") as f:
    f.write("ATGCGTACGT\nATG\nGTC\nTTTGGGCCCAAATTT\n")

conservées = 0
supprimées = 0
```

Le coeur du programme

```python
with open("brut.txt", "r") as lecture, open("filtre.txt", "w") as ecriture:
    for ligne in lecture:
        adn = ligne.strip()
        
        if len(adn) > 5:
            ecriture.write(adn + "\n")
            conservées += 1
        else:
            supprimées += 1
```

Affichage

```python
print(f"Nettoyage finie.")
print(f"{conservées} séquences ont été conservées.")
print(f"{supprimées} séquences ont été supprimées.")
```
