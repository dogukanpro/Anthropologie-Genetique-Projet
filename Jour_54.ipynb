Sujet du jour : le module json

```python
import json

profil_genetique = {
    "identifiant": "MAR-2026-01",
    "peuple": "Gallo-Romain",
    "marqueurs": ["A23", "B45", "T12"],
    "details": {
        "age_estime": 1850,
        "site": "Marseille Vieux-Port"
    }
}


with open("archive_genetique.json", "w", encoding="utf-8") as f:
    json.dump(profil_genetique, f, indent=4)
    print("Profil archivé en JSON !")

with open("archive_genetique.json", "r") as f:
    donnees_chargees = json.load(f)
    print(f"Lecture du site : {donnees_chargees['details']['site']}")
```
