```python
def calculer_ecart(freq_locale, freq_reference=0.50):
    ecart = (freq_locale - freq_reference)
    return ecart

# Cas 1 : On compare aux 0.50 par défaut 
print(f"Écart type Sapiens : {calculer_ecart(0.75):.2f}")

# Cas 2 : On compare à une référence spécifique
print(f"Écart spécifique Néandertal : {calculer_ecart(0.75, 0.08):.2f}")
```
