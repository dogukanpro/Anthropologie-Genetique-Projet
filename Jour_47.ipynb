Sujet : Importation du module csv et utilisation

```python
import csv

nom_f = "donnees_peuples.csv"

try:
    with open(nom_f, "r", encoding="utf-8") as f:
        lecteur = csv.DictReader(f)
        
        for ligne in lecteur:
            nom = ligne["Nom"]
            region = ligne["Region"]
            
            print(f"Le peuple {nom} habitait en {region}.")
except FileNotFoundError:
    print("Erreur : Fichier introuvable. Lance le code du Jour 46 d'abord !")
```
