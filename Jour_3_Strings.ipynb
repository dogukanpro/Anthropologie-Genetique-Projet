Une séquence d'ADN

```python
sequence = "atgcgtacgtagct"
```

Je la met en majuscule, pour correspondre aux normes

```python
sequence_majuscule = sequence.upper()
```

Pour connaitre la longueur

```python
longueur = len(sequence_majuscule)
```

Simulation d'une mutation

```python
sequence_mutee = sequence_majuscule.replace("G", "T")
```

Affichage

```python
print(f"La séquence avant mutation : {sequence_majuscule}")
print(f"La longueur est de : {longueur} bases")
print(f"La séquence après la mutation : {sequence_mutee}")
```

L'exemple avec une transformation d'ADN en ARN

```python
adn = "gatc"
print("En minuscules :", adn)

adn_propre = adn.upper()
print("En majuscules :", adn_propre)

arn = adn_propre.replace("T", "U")
print("Transformation en ARN :", arn)
```
