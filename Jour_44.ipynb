Sujet : Recherche efficace dans un fichier massif

```python
nom_f = "historique_analyses.txt"
```

Le cœur du programme

```python
try:
    with open(nom_f, "r") as f:
        
        for ligne in f:
            
            ligne_propre = ligne.strip()
            
#Affichage            
            if "Celtes" in ligne_propre:
                print(f"Trouvé : {ligne_propre}")

except FileNotFoundError:
    print("Erreur : Fichier introuvable.")

print("Recherche terminée.")
```
