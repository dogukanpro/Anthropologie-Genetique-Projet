La bibliothèque

```python
traducteur = {"AUG": "M", "UUU": "F", "GGC": "G", "UAA": "STOP"}
```

La séquence à traduire

```python
sequence_arn = ["AUG", "UUU", "GGC", "UAA"]
```

Le résultat finale

```python
proteine_finale = ""
```

Boucle for + affichage

```python
for codon in sequence_arn:
    acide = traducteur[codon]
    proteine_finale += acide + "-"


print(f"La protéine produite est : {proteine_finale}")
```
