```python
class Individu:
    def __init__(self, nom, peuple, score_ancetres):
        self.nom = nom
        self.peuple = peuple
        self.score = score_ancetres
    def se_presenter(self):
        print(f"[Fiche Bio] Nom: {self.nom} | Peuple: {self.peuple}")

    def verifier_purete(self):
        if self.score > 80:
            return "Haute conservation génétique"
        else:
            return "Profil mélangé"
```

```python
sujet_A = Individu("Alaric", "Goth", 85)
sujet_A.se_presenter() 
statut = sujet_A.verifier_purete()
print(f"Résultat analyse : {statut}")
```
