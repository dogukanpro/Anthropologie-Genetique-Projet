Sujet : Extraire un gene précis

```python
import re
adn_archaïque = "CCCGGG TATA ATGCGTACG GCG AAAAAA"

pattern = r"TATA\s*(.*)\s*GCG"
```

```python
recherche = re.search(pattern, adn_archaïque)

if recherche:
    sequence_extraite = recherche.group(1).strip()
    print(f"Gène extrait : {sequence_extraite}")
    print(f"Longueur : {len(sequence_extraite)} bases")
else:
    print("Structure du gène incomplète.")
```
