Sujet : Le module re et la fonction re.search()

```python
import re
adn_sequence = "GCTATAAAAGCTAGCTACGATCGAT"


motif = r"TATA[AT]+" 

resultat = re.search(motif, adn_sequence)

if resultat:

    print(f"Motif trouvé : {resultat.group()}")

    print(f"Position dans le gène : index {resultat.start()}")
else:
    print("Aucun signal de promoteur trouvé.")
```
