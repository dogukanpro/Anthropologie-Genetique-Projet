```python
mon_codon = "GGC"
```

Recherche dans le dictionnaire

```python
acide_amine = code_genetique[mon_codon]
```

Affichage

```python
print(f"Le codon {mon_codon} code pour l'acide aminé {acide_amine}.")
```
