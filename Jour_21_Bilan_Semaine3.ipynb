Initialisation, pour l'exemple j'ai pris la population : Anciens Asiatiques du Nord-Est (ANA Ancient Northeast Asian en anglais) et la population : Eurasiens du Nord (ANE Ancient North Eurasian en anglais)

```python
ana = "ATGCCG"
ane = "ATGCGG"
```

Création de la fonction pour analyser la mutation entre deux populations

```python
def analyser_mutation(seq1, seq2):
    total = 0
    for i in range(len(seq1)):
        if seq1[i] != seq2[i]
            total += 1
            print(print(f"Mutation trouvée à la position {i} : {seq1[i]} est devenue {seq2[i]}")
    return total
```

Affichage + utilisation de la fonction pour compter la mutation

```python
nb_mutations = analyser_mutation(ana, ane)
print(f"Bilan : {nb_mutations} mutations détectés.")
```
