Sujet : Importation du module csv et utilisation

```python
import csv

nom_f = "donnees_peuples.csv"
```

```python
with open(nom_f, "w", newline='') as f:
    writer = csv.writer(f)
    writer.writerow(["Nom", "Region", "Score"])
    writer.writerow(["Celtes", "Europe", "24.5"])
    writer.writerow(["Vikings", "Scandinavie", "12.8"])
```

```python
try:
    with open(nom_f, "r") as f:
        lecteur = csv.reader(f)
        
        next(lecteur)
        
        for ligne in lecteur:
            print(f"Peuple identifié : {ligne[0]}")

except FileNotFoundError:
    print("Fichier CSV introuvable.")
```
